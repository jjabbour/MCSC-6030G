# Numerical solution for the biharmonic equation         using Jacobi iterative method

The biharmonic equation with homogeneous Dirichlet boundary condition:

$\Delta^2 u = f(x,y) \; \textrm{on} \; \Omega \subseteq [0,1]\times[0,1]$      
$u(x,y) = 0  \;\; \textrm{on} \; \partial{\Omega}$       
$\Delta u = 0  \;\;\;\;\;\;\; \textrm{on} \; \partial{\Omega}$     

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import figure, show 
from mpl_toolkits.mplot3d import Axes3D
import fun_biharmonic_jacobi as bj
```

Initialize parameters for the problem and the Jacobi iteration method  

```python
n = 50
TOL = 1.0e-6
MAXITNS = 500000

h = 1.0/(n+1)
x = np.linspace(0,1,n+2)
(x,y) = np.meshgrid(x,x)
```

Given $f(x,y) = \pi^4(m^2 + p^2)^2\sin(m\pi x)\sin(p\pi y)$    
for $\Delta^2 u = f(x,y) \; \textrm{on} \; \Omega \subseteq [0,1]\times[0,1]$        
The exact solution is $ u(x,y) = \sin(m\pi x)\sin(p\pi y)$

```python
m = 1
p = 1
f = lambda x,y: (m**2 + p**2)**2*(np.pi**4)*np.sin(np.pi*m*x)*np.sin(np.pi*p*y)
F_RHS = np.zeros([n+2,n+2])
F_RHS[1:-1,1:-1] = (h**2)*f(x[1:-1,1:-1],y[1:-1,1:-1])
U_exact = np.sin(np.pi*m*x)*np.sin(np.pi*p*y)
```

The original problem Biharmonic equation can be transformed to two Poisson equation   
Jacobi iteration is used to solve the following system given an initial guess $v^{(0)}$

$\Delta v = f(x,y) \; \textrm{on} \; \Omega $     
$v = 0 \; \textrm{on} \; \partial{\Omega} $     
and    
$\Delta u = v \; \textrm{on} \; \Omega$     
$v = 0 \; \textrm{on} \; \partial{\Omega} $      


```python
U = np.zeros((n+2,n+2))
V = np.zeros((n+2,n+2))
delta_U = np.zeros([n,n])
delta_V = np.zeros([n,n])

for itn in range(MAXITNS):
    delta_V = bj.jacobi_sweep_2d(V,F_RHS)
    V[1:-1,1:-1] +=  delta_V
    delta_U = bj.jacobi_sweep_2d(U,h**2*V)
    U[1:-1,1:-1] +=  delta_U
    converged1 = bj.check_convergence(delta_U,U[1:-1,1:-1],TOL)
    converged2 = bj.check_convergence(delta_V,V[1:-1,1:-1],TOL)
    converged = (converged1) and (converged2)
    if (converged):
        print 'converged in', itn ,'iterations'
        break
```

## Figure 1: Plot of the numerical solution after obtaining convergence

```python
if not(converged):
    print "Jacobi iteration failed to converge after", itn, "iterations."
else:
    err = 1./n*(U_exact-U)
    fig1 = figure()
    Axes3D(fig1)
    plt.contourf(x,y,U,200)
    plt.colorbar()
    plt.show()    	
```

## Figure 2: Plot of the error between the numerical solution and the analytical solution

```python
print "The error between the exact and the numerical solution: ", \
np.linalg.norm(err.flatten())
fig2 = figure()
plt.contourf(x,y,err,200)
plt.colorbar()
plt.show()
```
